```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:
        mid = (high + low) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return 1, mid
    return 0, mid

def add_binary_search(arr,x):
    arr.append(x)
    arr.sort()
    return arr

def del_binary_search(arr,x):
    booly,index=binary_search(arr,x)
    if booly:
        arr=arr[:index]+arr[index+1:]
        return arr
```

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def tree_insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = tree_insert(root.right, key)
        else:
            root.left = tree_insert(root.left, key)
    return root

def tree_search(root, key):
    if root is None or root.val == key:
        return root
    if root.val < key:
        return tree_search(root.right, key)
    return tree_search(root.left, key)

def minValueNode(node):
    current = node
    while(current.left is not None):
        current = current.left 
    return current

def tree_deleteNode(root, key):
    if root is None:
        return root
    if key < root.val:
        root.left = tree_deleteNode(root.left, key)
    elif(key > root.val):
        root.right = tree_deleteNode(root.right, key)
    else:
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        temp = minValueNode(root.right)
        root.val = temp.val
        root.right = tree_deleteNode(root.right, temp.val)
    return root
```

```python
def fibonacci_search(arr, x):
    n=len(arr)
    fib2 = 0
    fib1 = 1
    fibM = fib2 + fib1
    while (fibM < n):
        fib2 = fib1
        fib1 = fibM
        fibM = fib2 + fib1
    offset = -1
    while (fibM > 1):
        i = min(offset+fib2, n-1)
        if (arr[i] < x):
            fibM = fib1
            fib1 = fib2
            fib2 = fibM - fib1
            offset = i
        elif (arr[i] > x):
            fibM = fib2
            fib1 = fib1 - fib2
            fib2 = fibM - fib1
        else:
            return 1,i
    if(fib1 and arr[offset+1] == x):
        return 1,offset+1
    return 0,offset+1

def add_fibonacci(arr,x):
    arr.append(x)
    arr.sort()
    return arr

def del_fibonacci(arr,x):
    booly,index=fibonacci_search(arr,x)
    if booly:
        arr=arr[:index]+arr[index+1:]
        return arr
```

```python
arr=[]
numbers=[4,-2,1,-3,8,-7,6,7,5]
for i in numbers:
    arr=add_binary_search(arr,i)
    print(arr)
for i in list(sorted(numbers)):
    print(binary_search(arr,i))
for i in numbers:
    arr=del_binary_search(arr,i)
    print(arr, i)
```
