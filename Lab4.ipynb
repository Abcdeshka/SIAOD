```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:
        mid = (high + low) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return 1, mid
    return 0, mid

def add_binary_search(arr,x):
    arr.append(x)
    arr.sort()
    return arr

def del_binary_search(arr,x):
    booly,index=binary_search(arr,x)
    if booly:
        arr=arr[:index]+arr[index+1:]
        return arr
```

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def tree_insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = tree_insert(root.right, key)
        else:
            root.left = tree_insert(root.left, key)
    return root

def tree_search(root, key):
    if root is None or root.val == key:
        return root
    if root.val < key:
        return tree_search(root.right, key)
    return tree_search(root.left, key)

def minValueNode(node):
    current = node
    while(current.left is not None):
        current = current.left 
    return current

def tree_deleteNode(root, key):
    if root is None:
        return root
    if key < root.val:
        root.left = tree_deleteNode(root.left, key)
    elif(key > root.val):
        root.right = tree_deleteNode(root.right, key)
    else:
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        temp = minValueNode(root.right)
        root.val = temp.val
        root.right = tree_deleteNode(root.right, temp.val)
    return root
```

```python
def fibonacci_search(arr, x):
    n=len(arr)
    fib2 = 0
    fib1 = 1
    fibM = fib2 + fib1
    while (fibM < n):
        fib2 = fib1
        fib1 = fibM
        fibM = fib2 + fib1
    offset = -1
    while (fibM > 1):
        i = min(offset+fib2, n-1)
        if (arr[i] < x):
            fibM = fib1
            fib1 = fib2
            fib2 = fibM - fib1
            offset = i
        elif (arr[i] > x):
            fibM = fib2
            fib1 = fib1 - fib2
            fib2 = fibM - fib1
        else:
            return 1,i
    if(fib1 and arr[offset+1] == x):
        return 1,offset+1
    return 0,offset+1

def add_fibonacci(arr,x):
    arr.append(x)
    arr.sort()
    return arr

def del_fibonacci(arr,x):
    booly,index=fibonacci_search(arr,x)
    if booly:
        arr=arr[:index]+arr[index+1:]
        return arr
```

```python
def interpolation_search(arr, lo, hi, x):
    if (lo <= hi and x >= arr[lo] and x <= arr[hi]):
        # Проверяем, не происходит ли деление на ноль
        if arr[hi] == arr[lo]:
            if arr[lo] == x:
                return 1, lo
            else:
                return 0, lo
        pos = lo + ((hi - lo) * (x - arr[lo]) // (arr[hi] - arr[lo]))
        if pos >= len(arr) or pos < 0:
            return 0, pos
        if arr[pos] == x:
            return 1, pos
        if arr[pos] < x:
            return interpolation_search(arr, pos + 1, hi, x)
        if arr[pos] > x:
            return interpolation_search(arr, lo, pos - 1, x)
    return 0, lo

def add_interpolation(arr,x):
    arr.append(x)
    arr.sort()
    return arr

def del_interpolation(arr,x):
    lo=0
    hi=len(arr)-1
    booly,index=interpolation_search(arr,lo,hi,x)
    if booly:
        arr=arr[:index]+arr[index+1:]
        return arr
```

```python
import random

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash(self, key):
        return key % self.size

    def rehash(self, oldhash):
        random.seed(oldhash)
        return (oldhash + random.randint(1, self.size - 1)) % self.size

    def put(self, key, value):
        slot = self.hash(key)

        if self.table[slot] is None or self.table[slot][0] == key:
            self.table[slot] = (key, value)
        else:
            newslot = self.rehash(slot)
            while self.table[newslot] is not None and self.table[newslot][0] != key:
                newslot = self.rehash(newslot)

            if self.table[newslot] is None or self.table[newslot][0] == key:
                self.table[newslot] = (key, value)

    def get(self, key):
        startslot = self.hash(key)

        data = self.table[startslot]
        if data is not None:
            if data[0] == key:
                return data[1]

        newslot = self.rehash(startslot)
        while self.table[newslot] is not None and self.table[newslot][0] != key:
            newslot = self.rehash(newslot)

        if self.table[newslot] is not None and self.table[newslot][0] == key:
            return self.table[newslot][1]
        else:
            return None


    
# Создаем хэш-таблицу размера 4
hash_table = HashTable(4)

hash_table.put(10, "Яблоко")
hash_table.put(17, "Банан")
hash_table.put(24, "Вишня")
hash_table.put(31, "Дыня")

print(hash_table.get(10)) 
print(hash_table.get(17)) 
print(hash_table.get(24)) 
print(hash_table.get(31))  
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash(self, key):
        return key % self.size

    def put(self, key, value):
        hashvalue = self.hash(key)

        if self.table[hashvalue] is None:
            self.table[hashvalue] = Node(key, value)
        else:
            node = self.table[hashvalue]
            while node.next:
                node = node.next
            node.next = Node(key, value)

    def get(self, key):
        hashvalue = self.hash(key)

        if self.table[hashvalue] is not None:
            node = self.table[hashvalue]
            while node:
                if node.key == key:
                    return node.value
                node = node.next

        return None
    
# Создаем хэш-таблицу размера 4
hash_table = HashTable(4)

hash_table.put(10, "Яблоко")
hash_table.put(17, "Банан")
hash_table.put(24, "Вишня")
hash_table.put(31, "Дыня")

print(hash_table.get(10)) 
print(hash_table.get(17)) 
print(hash_table.get(24)) 
print(hash_table.get(31))  
```

### Задание 3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.

```python
def is_safe(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    i, j = row, col
    while i < 8 and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1
    
    return True

def solve_queens(board, col):
    if col >= 8:
        return True
    
    for i in range(8):
        if is_safe(board, i, col):
            board[i][col] = 1
            if solve_queens(board, col + 1):
                return True
            board[i][col] = 0
    
    return False

def print_board(board):
    for i in range(8):
        for j in range(8):
            print(board[i][j], end=" ")
        print()

board = [[0 for _ in range(8)] for _ in range(8)]

if solve_queens(board, 0):
    print_board(board)
else:
    print("Решение не найдено.")

```
